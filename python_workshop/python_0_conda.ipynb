### Conda

**Conda** เป็นระบบการจัดการแพ็คเกจและการจัดการสภาพแวดล้อมในภาษา Python ช่วยให้เราสามารถติดตั้ง ใช้งาน และอัพเดตแพ็คเกจ ต่างๆ ได้ 

ดาวโหลด: https://docs.anaconda.com/free/miniconda/index.html

### Basic Commands

1. **Create a new environment**:
   ```
   conda create --name myenv
   ```
   You can also specify packages to install:
   ```
   conda create --name myenv numpy pandas
   ```

2. **Activate an environment**:
   ```
   conda activate myenv
   ```

3. **Deactivate the current environment**:
   ```
   conda deactivate
   ```

4. **List all environments**:
   ```
   conda env list
   ```

5. **List all packages in the active environment**:
   ```
   conda list
   ```

### Package Management

6. **Install a package**:
   ```
   conda install numpy
   ```
   You can specify the environment:
   ```
   conda install -n myenv numpy
   ```

7. **Update a package**:
   ```
   conda update numpy
   ```

8. **Uninstall a package**:
   ```
   conda remove numpy
   ```

### Environment Management

9. **Clone an environment**:
   ```
   conda create --name myenv_clone --clone myenv
   ```

10. **Remove an environment**:
    ```
    conda env remove --name myenv
    ```

11. **Export an environment**:
    This creates a YAML file with all the packages listed:
    ```
    conda env export > environment.yml
    ```

12. **Create an environment from a YAML file**:
    ```
    conda env create -f environment.yml
    ```

### Other Useful Commands

- **Search for a package**:
  ```
  conda search "numpy"
  ```

- **List all packages, channels, and dependencies for an environment**:
  ```
  conda list --explicit
  ```

- **Check for Conda updates**:
  ```
  conda update conda
  ```
